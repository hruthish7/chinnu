```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Hruthish Sreedhara - bitcoin.csv')
```

```python
data
```

```python
data1 = data[1:90]
```

```python
data1['gross profit or loss']=data1['Open']-data1['Close']
```

```python
data1
```

```python
import numpy as np
```

```python
x_values = data1['Date']
```

```python
x_values
```

```python
y_values = data1['gross profit or loss']
```

```python
y_values
```

```python
plt.figure(figsize=(25,10))
plt.xlabel('Date', fontsize = 20)
plt.ylabel('gross profit or loss',fontsize=20)
plt.title('bitcoin',fontsize=20)
plt.xticks(rotation=70,fontsize=12)
plt.yticks(fontsize=15)
plt.plot( x_values, y_values, marker='o', markerfacecolor='blue', markersize=5, color='skyblue', linewidth=1)
plt.savefig('foo1.png',transparent = True)
```
